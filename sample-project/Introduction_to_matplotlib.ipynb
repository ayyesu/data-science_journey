## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for workflow
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Set up your data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show (save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.set(title="Travis Nut Butter Store",
      xlabel="Butter",
      ylabel="Prices")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# Horizontal Bar
fig, ax = plt.subplots()
ax.set(title="Travis Nut Butter Store",
      xlabel="Prices",
      ylabel="Nut Butter")
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method)

* When plotting somthing quick, it is okay to use the pyplot
* When plotting something more advanced use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

#ax.set_xlim([45, 100])
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot data
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"])

# Customize plot
ax.set(title="Heart disease & Cholestrol rate",
      xlabel="age",
      ylabel="cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")

# Save figure
fig.savefig("over_50.png")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                              sharex=True)
# Add data to ax0 & ax1
scatter = ax0.scatter(over_50["age"], over_50["chol"], c=over_50["target"])
scatter_1 = ax1.scatter(over_50["age"], over_50["thalach"], c=over_50["target"])

# Customize ax0 & ax1
ax0.set(title="Heart Disease & Cholestrol level",
       ylabel="cholestrol")
ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="age",
       ylabel="Max Heart Rate")

# Add legend
ax0.legend(*scatter.legend_elements(), title="Target")
ax1.legend(*scatter_1.legend_elements(), title="Target")

# Add horizontal line
ax0.axhline(over_50["chol"].mean(), linestyle="--");
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# save the figure
fig.savefig("heart_disease_analysis.png");
```
