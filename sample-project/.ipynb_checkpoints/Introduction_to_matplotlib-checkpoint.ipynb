## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for workflow
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Set up your data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show (save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Making figures with numpy arrays

We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np 
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to create scatter plots
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.set(title="Travis Nut Butter Store",
      xlabel="Butter",
      ylabel="Prices")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# Horizontal Bar
fig, ax = plt.subplots()
ax.set(title="Travis Nut Butter Store",
      xlabel="Prices",
      ylabel="Nut Butter")
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrame

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range("1/1/2024", periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2].astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2022", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
