Now, if we want to find the maximum or minimum from the rows or the columns then we have to add 0 or 1. See how it works: 

```python
arr = np.array([[3, 2, 8],
         [5, 0, 8],
         [23, 89, 8]])
arr
```

```python
max_col_element = np.max(arr, 0)
min_row_element = np.min(arr, 1)
```

```python
print("Max element on the columns is: ", max_col_element)
print("Minimun elemnt on the rows is: ", min_row_element)
```

Array Sorting

```python
arr = np.array([
    [3, 2, 7],
    [86, 4, 20],
    [2, 12, 4]
])
arr
```

Sorted of the column axis

```python
arr1 = np.sort(arr, 0)
arr1
```

Sorted on the row axis

```python
arr2 = np.sort(arr, 1)
arr2
```

Sorted on no axis

```python
arr3 = np.sort(arr, axis=None)
arr3
```

We can choose which type of sorting we want perform using the kind parameter inside the sort function. We can either choose between quicksort, mergesort and heapsort.

```python
arr4 = np.sort(arr, axis=None, kind='mergesort')
arr4
```

Array Merging

```python
first = np.array([3, 9, 5])
second = np.array([43, 2, 6])
```

```python
np.vstack((first, second))
```

```python
np.hstack((first, second))
```

```python
third = np.array([[2, 4], [6, 8]])
fourth = np.array([[3, 5], [7, 9]])
```

```python
np.concatenate((third, fourth), axis=0)
```

```python
np.concatenate((third, fourth), axis=1)
```
