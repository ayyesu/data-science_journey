```python
import requests
```

Accessing HTML content of webpage

```python
url = "http://www.values.com/inspirational-quotes"
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'}
r = requests.get(url=url, headers=headers)
print(r.content)
```

Parsing the HTML

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

Searching and navigating through the parse

```python
quotes = []

table = soup.find('div', attrs={'id': 'all_quotes'})
print(table)
```

```python
for row in table.findAll('div', attrs={'class': 'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['image'] = row.img['src']
    quote['lines'] = row.img['alt'].split(' #')[0]
    quote['author'] = row.img['alt'].split(' #')[1]
    quotes.append(quote)
```

```python
# Saving to a csv file
import csv

filename = 'inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['theme','url','image','lines','author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
